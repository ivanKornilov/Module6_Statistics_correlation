```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

Таблица сопряженности:


Оснавня и альтернативная гипотезы, сформулироанные на математическом языке:


Задача 1
Изучается зависимость выбираемой абитуриентами специальности от пола. Как показал опрос, из 355 абитуриентов мужского пола 197 собирались поступать на специальность, относящуюся к точным наукам, а 158 — на специальность, относящуюся к гуманитарным наукам. Среди абитуриентов женского пола точные науки выбрали 102 человека, гуманитарные — 221 человек. Что можно сказать о зависимости между полом абитуриента и выбором специальности?

Таблица сопряженности:


```python
obs = np.array([[197,158],[102,221]])
obs
# таблица сопряженности
```

# проверка стат. гипотезы


```python
stats.chi2_contingency(obs)
# возвращает значение статистики хи-квадрат
# p-значение (при 5% если близко к 0 отрицать не можем)
# ожидаемые частоты
```

```python
# т.к. p-значение показателя близко к 0, не можем отрицать Нулевую гипотезу, т.е. Зависимости от пола выбора специальности нет
```

# сила зависимости:
от 0 до 1
0 - признакии независимы
1 - зная один признак можно в точности предсказать другой


```python
chi2 = stats.chi2_contingency(obs)[0] # значение статистики хиквадрат с крышкой
n = np.sum(obs) 
k = obs.shape[0]
m = obs.shape[1]

tchuprow = np.sqrt(chi2 / (n * np.sqrt((k-1) * (m-1))))
cramer = np.sqrt(chi2 / (n*(np.minimum(k,m) - 1)))
pearson = np.sqrt(chi2 / (n + chi2))

print(tchuprow) # коэфф чупрова
print(cramer) #коэфф крамера
print(pearson) #коэфф пирсона
```

```python
# если измерять силу зависимости между полом и выбором специальности, то
# по шкале от 0 до 1,  примерно оценивается, как 0,2 т.е. крайне слабая положительная
```

# коэффициент контингенции 
сила двустронней зависимости между признаками
от -1 до 1
Если 0 - признаки независимы
1 - есть А есть В
-1 - есть А нет В



```python
F_ = np.sqrt(obs[0][0] + obs[1][0])*(obs[0][1] + obs[1][1])*(obs[0][0] + obs[0][1])*(obs[1][0] + obs[1][1])
F = (obs[0][0]*obs[1][1]-obs[0][1]*obs[1][0])/F_
F
```

```python
# признаки независимы
```

# коэффициент Юла
показывает силу односторонней связи

сила односторонней связимежду признаками 
от -1 до 1 
Если 0 - признаки независимы 
1 - есть А есть В   ИЛИ есть В есть А 
-1 - есть А нет В   ИЛИ есть В нет А



```python
Q_ = obs[0][0]*obs[1][1] + obs[0][1]*obs[1][0]
Q = (obs[0][0]*obs[1][1]-obs[0][1]*obs[1][0])/Q_
Q
```
