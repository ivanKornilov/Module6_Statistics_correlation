```python
chi2 = stats.chi2_contingency(obs)[0] #     
n = np.sum(obs) 
k = obs.shape[0]
m = obs.shape[1]

tchuprow = np.sqrt(chi2 / (n * np.sqrt((k-1) * (m-1))))
cramer = np.sqrt(chi2 / (n*(np.minimum(k,m) - 1)))
pearson = np.sqrt(chi2 / (n + chi2))

print(tchuprow) #  
print(cramer) # 
print(pearson) # 
```
